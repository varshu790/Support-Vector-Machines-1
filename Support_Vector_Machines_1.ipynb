Q1. What is the mathematical formula for a linear SVM?

ANS-The mathematical formula for a linear Support Vector Machine (SVM) involves finding the decision boundary represented by a hyperplane that best separates the classes in the dataset. For a linearly separable dataset, the formula is:

\[ \mathbf{w} \cdot \mathbf{x} + b = 0 \]

where:
- \(\mathbf{w}\) is the weight vector perpendicular to the hyperplane.
- \(\mathbf{x}\) is the input feature vector.
- \(b\) is the bias term.

The decision boundary is formed by the set of points where \(\mathbf{w} \cdot \mathbf{x} + b = 0\). The sign of \(\mathbf{w} \cdot \mathbf{x} + b\) determines the class to which a point belongs:
- If \(\mathbf{w} \cdot \mathbf{x} + b > 0\), the point belongs to one class.
- If \(\mathbf{w} \cdot \mathbf{x} + b < 0\), the point belongs to the other class.

During the training of the SVM, the objective is to find \(\mathbf{w}\) and \(b\) that maximize the margin (distance) between the hyperplane and the closest points (support vectors) of each class while minimizing classification errors. This is typically solved as an optimization problem using techniques like quadratic programming or convex optimization.

Q2. What is the objective function of a linear SVM?

ANS-The objective function of a linear Support Vector Machine (SVM) aims to find the optimal hyperplane that maximizes the margin between classes while minimizing the classification error. This objective function is often formulated as a constrained optimization problem.

For a linearly separable dataset, the objective function of a linear SVM can be expressed as:

\[ \min_{\mathbf{w}, b} \frac{1}{2} \|\mathbf{w}\|^2 \]

subject to the constraints:

\[ y_i(\mathbf{w} \cdot \mathbf{x}_i + b) \geq 1 \text{ for } i = 1, 2, \dots, N \]

where:
- \(\mathbf{w}\) is the weight vector perpendicular to the hyperplane.
- \(b\) is the bias term.
- \(\mathbf{x}_i\) represents the feature vector of the \(i\)th training sample.
- \(y_i\) is the class label of the \(i\)th training sample (-1 or 1 for binary classification).
- \(N\) is the number of training samples.

The objective function aims to minimize \(\frac{1}{2} \|\mathbf{w}\|^2\), which corresponds to maximizing the margin while the constraints \(y_i(\mathbf{w} \cdot \mathbf{x}_i + b) \geq 1\) ensure that the training samples are correctly classified and lie outside a certain margin from the decision boundary.

This problem can be solved using optimization techniques like quadratic programming or convex optimization to find the optimal \(\mathbf{w}\) and \(b\) that define the separating hyperplane.

Q4. What is the role of support vectors in SVM Explain with example

ANS- Support vectors play a crucial role in Support Vector Machines (SVMs) as they define the decision boundary or hyperplane. In a nutshell, they are the data points closest to the hyperplane and have a direct influence on its position. These vectors support or define the margin, hence the name "support vectors."

Here's an example to illustrate their role:

Imagine a simple binary classification problem where you have two classes, represented by circles (○) and crosses (✗), plotted on a 2D plane:

```
     ✗       ○
    ✗    ○
   ✗   ○
  ✗○  
  -------------------
```

In this case, the circles and crosses represent two classes that are linearly separable. To classify these points, you can draw a separating hyperplane. But what makes SVM unique is that it doesn't just find any separating line; it finds the one that maximizes the margin between classes.

When an SVM is trained on this dataset, it identifies the support vectors—these are the data points that lie closest to the decision boundary (the hyperplane). In this scenario, they might be the points directly on the edge or within the margin.

Let's say the support vectors are two crosses (✗) and two circles (○) marked below:

```
     ✗       ○
    ✗S   ○S
   ✗   ○
  ✗○S
  -------------------
```

These support vectors directly influence the position of the decision boundary. They determine the orientation and location of the hyperplane. The other data points farther away from the decision boundary do not affect the boundary's position in the current scenario.

During the training process, SVM identifies and focuses only on these support vectors to define the optimal hyperplane, making the algorithm memory efficient and computationally faster even with large datasets.

Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in
SVM?

ANS-Certainly! Let's use some examples and graphs to illustrate these concepts in Support Vector Machines (SVMs).

### Hyperplane:

A hyperplane in a 2D space is a line that separates two classes. In a 3D space, it's a plane, and in higher dimensions, it's a generalized separating surface. Consider a simple dataset where two classes, represented by red and blue points, are linearly separable:

![Hyperplane](https://i.ibb.co/gDZFzL2/hyperplane.png)

In the above graph, the straight line (hyperplane) separates the red and blue points.

### Margin:

The margin is the region surrounding the hyperplane that separates the classes. In a hard-margin SVM, this region is clear of any data points. In a soft-margin SVM, some points may be allowed within the margin.

![Margin](https://i.ibb.co/FHDdWL5/margin.png)

The area between the dashed lines is the margin. The solid line is the hyperplane. In a hard-margin SVM, no data points lie within this region.

### Soft Margin:

When the data is not perfectly separable, allowing some points to be inside the margin can be necessary for a better classification. A soft-margin SVM allows for a margin violation (data points within the margin) by introducing a penalty for misclassifications.

![Soft Margin](https://i.ibb.co/BtcHqKc/soft-margin.png)

The above image shows a scenario where a soft-margin SVM allows for a few misclassified points within the margin.

### Hard Margin:

Contrarily, a hard-margin SVM demands that all data points be correctly classified and lie outside the margin. This works only when the data is perfectly separable by a hyperplane.

![Hard Margin](https://i.ibb.co/1XsK9tQ/hard-margin.png)

In the graph above, no data point is allowed within the margin in a hard-margin SVM, and all points are correctly classified.

These illustrations demonstrate how the concept of the hyperplane, margin, soft margin, and hard margin operates in the context of SVMs, emphasizing the trade-off between finding a wider margin and allowing some misclassifications (soft margin) or demanding a perfect classification (hard margin).

````python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import numpy as np
import matplotlib.pyplot as plt

# Load Iris dataset
iris = datasets.load_iris()
X = iris.data[:, :2]  # Considering only the first two features for visualization
y = iris.target

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a linear SVM classifier
def train_linear_svm(C_value):
    svm_classifier = SVC(kernel='linear', C=C_value)
    svm_classifier.fit(X_train, y_train)
    return svm_classifier

# Function to plot decision boundaries
def plot_decision_boundary(svm, X, y):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
    Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.4)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=20, edgecolors='k')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('Decision Boundary')

# Train SVM and plot decision boundary
def train_and_plot(C_value):
    svm = train_linear_svm(C_value)
    plt.figure(figsize=(8, 6))
    plot_decision_boundary(svm, X_train, y_train)
    plt.show()
    return svm

# Function to calculate accuracy on test set
def calculate_accuracy(svm, X_test, y_test):
    accuracy = svm.score(X_test, y_test)
    return accuracy

# Train SVM with different values of C and evaluate performance
C_values = [0.1, 1, 10, 100]
for C_val in C_values:
    svm_model = train_and_plot(C_val)
    accuracy = calculate_accuracy(svm_model, X_test, y_test)
    print(f"Accuracy with C={C_val}: {accuracy:.4f}")
````

````python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

# Load Iris dataset
iris = datasets.load_iris()
X = iris.data[:, :2]  # Considering only the first two features for simplicity
y = iris.target

# Change labels to -1 and 1 for SVM formulation
y[y == 0] = -1

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# SVM from scratch using Sequential Minimal Optimization (SMO)
class LinearSVM:
    def __init__(self, C=1.0, tol=0.001, max_iter=1000):
        self.C = C
        self.tol = tol
        self.max_iter = max_iter
        self.alpha = None
        self.bias = None

    def fit(self, X, y):
        num_samples, num_features = X.shape
        self.alpha = np.zeros(num_samples)
        self.bias = 0.0
        for _ in range(self.max_iter):
            alpha_changed = 0
            for i in range(num_samples):
                prediction = np.dot(X[i], X.T).dot(y * self.alpha) + self.bias
                error_i = prediction - y[i]

                if (y[i] * error_i < -self.tol and self.alpha[i] < self.C) or \
                   (y[i] * error_i > self.tol and self.alpha[i] > 0):

                    j = np.random.choice(np.delete(np.arange(num_samples), i))
                    prediction = np.dot(X[j], X.T).dot(y * self.alpha) + self.bias
                    error_j = prediction - y[j]

                    alpha_i_old, alpha_j_old = self.alpha[i], self.alpha[j]

                    if y[i] != y[j]:
                        L = max(0, self.alpha[j] - self.alpha[i])
                        H = min(self.C, self.C + self.alpha[j] - self.alpha[i])
                    else:
                        L = max(0, self.alpha[i] + self.alpha[j] - self.C)
                        H = min(self.C, self.alpha[i] + self.alpha[j])

                    if L == H:
                        continue

                    eta = 2 * np.dot(X[i], X[j]) - np.dot(X[i], X[i]) - np.dot(X[j], X[j])
                    if eta >= 0:
                        continue

                    self.alpha[j] -= y[j] * (error_i - error_j) / eta
                    self.alpha[j] = max(min(self.alpha[j], H), L)

                    if abs(self.alpha[j] - alpha_j_old) < 1e-5:
                        continue

                    self.alpha[i] += y[i] * y[j] * (alpha_j_old - self.alpha[j])

                    b1 = self.bias - error_i - y[i] * (self.alpha[i] - alpha_i_old) * np.dot(X[i], X[i]) - \
                         y[j] * (self.alpha[j] - alpha_j_old) * np.dot(X[i], X[j])
                    b2 = self.bias - error_j - y[i] * (self.alpha[i] - alpha_i_old) * np.dot(X[i], X[j]) - \
                         y[j] * (self.alpha[j] - alpha_j_old) * np.dot(X[j], X[j])

                    if 0 < self.alpha[i] < self.C:
                        self.bias = b1
                    elif 0 < self.alpha[j] < self.C:
                        self.bias = b2
                    else:
                        self.bias = (b1 + b2) / 2

                    alpha_changed += 1

            if alpha_changed == 0:
                break

    def predict(self, X):
        return np.sign(np.dot(X, X_train.T).dot(y_train * self.alpha) + self.bias)

# Train SVM from scratch
svm_scratch = LinearSVM(C=1)
svm_scratch.fit(X_train, y_train)

# Train SVM using scikit-learn for comparison
svm_sklearn = SVC(kernel='linear', C=1)
svm_sklearn.fit(X_train, y_train)

# Prediction from scratch and scikit-learn implementations
predictions_scratch = svm_scratch.predict(X_test)
predictions_sklearn = svm_sklearn.predict(X_test)

# Calculate accuracy
accuracy_scratch = np.mean(predictions_scratch == y_test)
accuracy_sklearn = np.mean(predictions_sklearn == y_test)

print(f"Accuracy from scratch: {accuracy_scratch:.4f}")
print(f"Accuracy using scikit-learn: {accuracy_sklearn:.4f}")
````
